## Log Levels
Here are the standard log levels supported by Python `logging` library.
* DEBUG
* INFO
* WARNING (Default)
* ERROR
* CRITICAL
* For all the log levels, logging library provide a function. For example we can log **INFO** messages using `logging.info`

Based up on the log level used only the log messages generated by higher levels than the used one will be logged. With **WARNING** as default, only the messages generated by `logging.error` and `logging.critical` will be logged.

```python
import logging
logging.warning('This is Warning!')  # You will get this message logged by default
logging.info('Just Informational')  # No message will be logged as we are using default log level
```

## Log Format

```python
import logging, sys
logging.basicConfig(stream=sys.stdout, format='%(levelname)s:%(message)s', level=logging.INFO)
logging.debug('Debugging')
logging.info('Informational, typically used before important functions')
logging.warning('Acceptable abnormal behavior is logged as warnings')
logging.error('Typically used as part of the exceptions')
```

## Logging to File

Here is a basic example of logging to file

```python
import logging
logging.basicConfig(format='%(asctime)s %(message)s')
logging.warning('is when this event was logged.')
```
